# EDA

Load OHLCV from `data/raw/` (see `data/download.py`), then inspect returns and rolling volatility.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

ROOT = Path.cwd().parent if Path.cwd().name == "eda" else Path.cwd()
csv_path = ROOT / "data" / "raw" / "SPY.csv"
import numpy as np

df = pd.read_csv(csv_path)
df.columns = [str(c).strip().lower() for c in df.columns]
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")
df["logret"] = np.log(df["close"] / df["close"].shift(1))
df["vol20"] = df["logret"].rolling(20).std()
df.tail()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df["date"], df["close"])
ax.set_title("Close")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 3))
sns.histplot(df["logret"].dropna(), bins=80, kde=True, ax=ax)
ax.set_title("Distribution of log returns")
plt.tight_layout()
plt.show()
```
